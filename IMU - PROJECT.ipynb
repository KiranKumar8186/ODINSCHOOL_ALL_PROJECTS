# For data In Stock:

```python
# by using get_group function to getting separation of 'In Stock' from 'status'

df2 = df1.get_group("In Stock")
df2
```

## 1. Which are the top Manufacturers with price, quantity?

```python
# finding the top manufacturers with price, quantity
# so, multiplying the quantity and price and creating a new column
# In my scenario top manufacturer will be desided based on the total price of manufacturer

df2["total_price"] = df2["qty"]*df["price"]
df2
```

```python
# arranging the total_price in high to low
# then using head() getting 1st 5 valuess from the obtained data

df2.sort_values("total_price", ascending=False).head()
```

## 2. Which are the most profitable products?

```python
# most profitable products will be the high qty with high price of the products 
# by using max() function getting the high total_price of the instock data

df2[df2.total_price==df2.total_price.max()]
```

```python
# total_price of the manufacturers are in high to low order wise below.
# from those products manufacturers are getting more money
# these are top 5 most profitable products

df2.sort_values("total_price", ascending=False).head()
```

## 3. Which is the product where you have seen a decrease in the price?

```python
# using sort_values to getting the decrease in the price of the products
# ascending = False is used to get the values in high to low

a = df2.sort_values("price", ascending = False)
a
```

```python
# alternate method
# from this function also we can find that which product price is decreasing
# in this function I've used description in first, so we are getting the order of the products
# then after getting the order we can get by comparing the 10 days data with price to the products
# then we can sort that data and get the price decreasing product

df2.sort_values(by = ['description'], ascending = [False])
```

## 4. Which is the product where you have seen an increase in the price?

```python
# using sort_values to getting the decrease in the price of the products
# ascending = True is used to get the values in high to low

b = df2.sort_values("price", ascending = True)
b
```

```python
# alternate method
# from this function also we can find that which product price is increasing
# in this function I've used description in first, so we are getting the order of the products
# then after getting the order we can get by comparing the 10 days data with price to the products

df2.sort_values(by = ['description'], ascending = [False])
```

## 5. Which is the top category by price and quantity?

```python
# here if we can get the max price value of product
# then that product category will be the top 
# so, here finding the max price value product
# '521' is the top category based on price

df2[df2.price==df2.price.max()]
```

```python
# here if we can get the max quantity of product
# then that product category will be the top 
# so, here finding the max quantity of product
# '1221' is the top category based on quantity

df2[df2.qty==df2.qty.max()]
```

## 6. Which products are aging?

```python
# Taking the 1st day data
# arranginf quantity of products in high to low
# by using min( ) to day one detailes and sort_values to get quantity order in high to low 

df2_qty = df2[df2.sku_create_date==df2.sku_create_date.min()]
df2_qty.sort_values("qty", ascending=False).head(10)
```

```python
# from below and above output
# now we can compare the the data present in both dates
# so, from the output if the quantity of the products not changes 
# then those products are aging
# from the output '1221' category of the products are still in stock. No sold much. So those are aging products
```

```python
# Like above here Taking the last day data
# arranginf quantity of products in high to low
# by using max( ) to last day detailes and sort_values to get quantity order in high to low

df2_qty2 = df2[df2.sku_create_date==df2.sku_create_date.max()]
df2_qty2.sort_values("qty", ascending=False).head(10)
```

## 7. How to Decrease Product aging?

###### Tips to decrease Product aging :

* When product aging is noticed you have identified the aged stock from the aging inventory report and the age buckets of this stock, you will start marking down your products based on the age bucket they belong to.


* It is very important to apply the discount gradually and not just give a flat discount on everything, because when you give higher discount on more aged stocks you will make it more attractive to customer and this automatically ensures clearing those stocks faster than the newer ones.


* This gradual discount also ensures you are making the most money out of your stock, by not giving unnecessary discounts on stock that is not too old yet.


* Giving the offers like buy one and get one. Then aged products are going to decrease. And that offer will be more atrctive to customers. 'And picking the right price to the product is most importent'. Customer mindset is like that product is worth to the price or not. So in starting we can just expect normal profit on product, then we can gradually increase the price. It'll decrease product aging definitely. 


* social media utilization will be the great option to promote the product and decreasing of the product aging. Social media is a free space where we can connect with your customers daily. If your customer is using social media, you need to be using social media. This will keep you top of mind when they coonsider what they want or need to buy.
